## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
db.list_collection_names()
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question:

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count_hygiene_20 = establishments.count_documents(query)
print(f"Number of establishments with hygiene score 20: {count_hygiene_20}")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
df_hygiene_20 = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df_hygiene_20)}")

# Display the first 10 rows of the DataFrame
df_hygiene_20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "AddressLine3": {"$regex": "London", "$options": "i"},  # Case-insensitive matching
    "RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
count_high_rating_london = establishments.count_documents(query)
print(f"Number of establishments in London with RatingValue >= 4: {count_high_rating_london}")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
df_high_rating_london = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df_high_rating_london)}")

# Display the first 10 rows of the DataFrame
df_high_rating_london.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query_top_5 = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

sort_top_5 = [("scores.Hygiene", 1)]  # Sort by lowest hygiene score

# Print the results
top_5_results = establishments.find(query_top_5).sort(sort_top_5).limit(5)
for result in top_5_results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
df_top_5 = pd.DataFrame(list(establishments.find(query_top_5).sort(sort_top_5).limit(5)))

# Display the first 5 rows of the DataFrame
df_top_5.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline_hygiene_0 = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Print the number of documents in the result
result_hygiene_0 = list(establishments.aggregate(pipeline_hygiene_0))
pprint(result_hygiene_0)

# Print the first 10 results
pprint(result_hygiene_0[:10])
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene_0 = pd.DataFrame(result_hygiene_0)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df_hygiene_0)}")

# Display the first 10 rows of the DataFrame
df_hygiene_0.head(10)
```
